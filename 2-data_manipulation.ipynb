```python
import pandas as pd
```

```python
# Carga de archivo "Datos Maestros VF.xlsx", en la hoja "Master Data Oficial"
nombre = 'Nombre visible Agente'
agente = 'AGENTE (OFEI)'
central = 'CENTRAL (dDEC, dSEGDES, dPRU…)'
tipo = 'Tipo de central (Hidro, Termo, Filo, Menor)'

filename = r"Datos3\Datos Maestros VF.xlsx"
sheet = r"Master Data Oficial"

columns =[nombre, agente, central, tipo]

df1 = pd.read_excel(filename, sheet_name=sheet, usecols=columns)
df1
```

```python
# Para determinar si hay valores emgesa mal escritos
df1[agente].unique()
```

```python
# Seleccionar los registros que pertenecen al agente EMGESA ó EMGESA S.A. y adicionalmente que el Tipo de Central sea ‘H’ o ‘T’.
df1[(df1[agente] == "EMGESA S.A.") & 
   ((df1[tipo] == "H") |
    (df1[tipo] == "M"))]
```

```python
# Carga de archivo "dDEC1204.TXT"
filename = r"Datos3\dDEC1204.TXT"
columns = [central] + ["DATA_"+str(i) for i in range(1, 25)]

df2 = pd.read_csv(filename, encoding="latin1", names=columns)
df2
```

```python
# Realizando el merge de los dos data sets por Central.
resultado = pd.merge(df1, df2, on=central, how="inner")
resultado
```

```python
# Left join para encontrar filas en df1 que no están en df2
left_join = pd.merge(df1, df2, on=central, how='left', indicator=True)
not_in_df2 = left_join[left_join['_merge'] == 'left_only']
print("Centrales de 'Datos Maestros VF.xlsx' que no están en 'dDEC1204.TXT':")
print(not_in_df2.drop(columns='_merge')[central].unique())
```

```python
# Right join para encontrar filas en df2 que no están en df1
right_join = pd.merge(df1, df2, on=central, how='right', indicator=True)
not_in_df1 = right_join[right_join['_merge'] == 'right_only']
print("Centrales de 'dDEC1204.TXT' que no están en 'Datos Maestros VF.xlsx':")
print(not_in_df1.drop(columns='_merge')[central].unique())
```

```python
# Calcular la suma horizontal de todas las horas para cada planta
resultado["TOTAL"] = resultado.filter(like="DATA_").sum(axis=1)
resultado[[central, "TOTAL"]]
```

```python
# Seleccionar solamente los registros de las plantas cuya suma horizontal sea mayor que cero
resultado_filtrado = resultado[resultado["TOTAL"] > 0][[central, "TOTAL"]]
resultado_filtrado
```

```python
resultado_filtrado.to_csv(r'segundo_ejercicio.csv', index=False)
```
